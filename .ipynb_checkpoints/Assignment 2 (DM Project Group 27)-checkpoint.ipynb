### Importing all necessary libraries

```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

### Inserting the data

```python
# Inserting the data
data = pd.read_csv('data/googleplaystore.csv')
data
```

### Cleaning the data
#### Dropping columns

```python
# Cleans the data using the dropna() and drop_duplicates() function from pandas
# The dropna() function will remove rows or columns which contain missing values
data_clean = data.dropna()
data_clean
```

```python
# The drop_duplicates() function will remove any duplicate data
data_clean = data_clean.drop_duplicates()
data_clean
```

```python
data_clean = data_clean.reset_index(drop=True)
data_clean
```

#### Converting values

```python
# Converting the 'Reviews', 'Price' columns to appropriate datatypes
# 'Reviews' column to Int
data_clean.Reviews = pd.to_numeric(data_clean.Reviews, errors='coerce')
data_clean.Reviews
```

```python
# 'Price' column to Float
data_clean.Price = data_clean.Price.str.replace('$','')
data_clean.Price = pd.to_numeric(data_clean.Price, errors='coerce')
data_clean.Price
```
